```python
#Webscraping de 1 pagina

pagina = ("https://www.letras.com/sex-pistols/")
pagina_url = requests.get(pagina)
print(pagina_url)
soup = BeautifulSoup(pagina_url.content, 'html.parser')
#mucho mas largo el proceso que con la libreria Rvest en Rstudio para obtener el <DOCTYPE HTML> = soup

#extraer los links de todas las canciones
links_letras = soup.find_all('div', class_='cnt-list--alp' )
shareurl_node = soup.find_all(attrs={'data-shareurl': True})

lista_paginas_canciones = [] #creo una variable vacia para llenar con el proceso for 
for x in shareurl_node: # para x que seria el nombre cualquier da la funcion y el nombre de lo que queiro recorrer
    lista_paginas_canciones.append(x['data-shareurl'])#
print (lista_paginas_canciones)



canciones_sex_pistol = []
for x in lista_paginas_canciones:
    pagina = requests.get(x)
    soup = BeautifulSoup(pagina.content, 'html.parser')
    letras = soup.find_all('div', class_='cnt-letra')
    for letra in letras:
        canciones_sex_pistol.append(letra.prettify().replace('<br/>', '\n'))
print(canciones_sex_pistol)
```

```python
proceso1 = []
for letra in canciones_sex_pistol:
    letra_procesada = re.sub(r'<div class="cnt-letra">|<p>|</p>|</div>', ' ', letra)
    proceso1.append(letra_procesada)

print(proceso1[1])
```
